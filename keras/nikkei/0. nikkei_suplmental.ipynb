```python
# this is trying to re-implment nikkei.py
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("nikkei.csv", header=None)
```

```python
print(df)
```

```python
# Only Stock prices
prices = df[1]
print(prices)
```

```python
# Is price up or down?
for i in range(len(prices)):
    if i > 0:
        print((prices[i] - prices.iloc[i - 1]) > 0)
```

```python
def upOrDown(df):
    ret = []
    for i in range(len(df)):
        if i > 0:
            ret.append(1 if (df.iloc[i] - df.iloc[i - 1]) > 0 else 0)
    return ret
```

```python
upOrDown(prices)
```

```python
# Trying to caliculate loss for upOrDown
def upOrDownLoss(a, b):
    ret = 0
    for i in range(len(a)):
        ret += pow(a[i] - b[i], 2)
    return ret / len(a)
```

```python
x = upOrDown(prices)
y = upOrDown(prices)
upOrDownLoss(x, y)
```
